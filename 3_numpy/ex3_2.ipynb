## NumPy

Use the temp.txt file and load it in an array. Construct a histogram composed of 5 intervals, creating an array of the frequencies of value occurance as the bars of these intervals.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Laod data
data = np.loadtxt('intel_lab/temp.txt')

# Flatten 2D into 1D
flattened = data.ravel()  # shape (632*48)

# Histogram with 5 intervals
freq, bin_edges = np.histogram(flattened, bins=5)
print("Histogram frequencies:", freq)
print("Bin edges:", bin_edges)

# Plot the histogram
plt.hist(flattened, bins=5, edgecolor='black')
plt.title("Histogram of Temperature Values (5 intervals)")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```
