```python
import pandas as pd

df = pd.read_table("intel_raw.txt",\
                  sep=" |\t", parse_dates = True)
```

```python
df.info(verbose = True, show_counts = True)
```

```python
# Subset is which columns to check for NaN
df.dropna(subset = ["temperature", "humidity", "light", "voltage"], how = "all", inplace=True, axis = 0 )

# Showing only the rows and columns of the dataframe that have at least one NaN
df[df.isnull().all(axis=1)]
```

```python
# Fill NaN values using linear interpolation
# I used interpolate because in the case that one value is missing, I get the average of the previous and next non-NaN values.
# In the case that more values are missing, I take the linear interpolation between them.
df = df.interpolate(method='linear', axis=0, limit_direction='both')

df.info(verbose = True, show_counts = True)
```
